# Serie de Fourier de una función a trozos

Nuestra función esta definida por:


$$
f(x) =
\begin{cases}
1 & \text{si } 0 \leq x < L/2 \\
0 & \text{si } L/2\leq x \leq L
\end{cases}
$$



Obteniendo el términos para la serie de Fourier:
$$b_n=\frac{-2}{n\pi}[cos(\frac{n\pi}{2})-1]$$

Con eso, obtenemos la siguiente serie:


$$g(x)=\sum_{n=1}^{\infty}(\frac{-2}{(n\pi)}[ \cos(\frac{n\pi}{2})-1]e^{-n^2t}).$$

A continuación graficaremos nuestra función y nuestra serie de Fourier

```python
from pylab import *
from math import pi
from numpy import linspace,zeros
```

```python
def recta(x):
    y = zeros(len(x))
    for i in range(len(x)):
        if x[i]  <= pi/2:
          y[i] = 1
        elif x[i] > pi/2:
          y[i] = 0
    return y
```

```python
x = linspace(-3,3,1000) #DUDA
f = recta(x)
plot(x,f,label='$u(x,0)$', color="DeepPink")
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Gráfica de nuestra función definida a trozos, condición incial")
xlim([-2,3]),ylim([-2.5,2.5]),grid(),legend()
show()
```

Definimos nuestra serie de Fourier y la graficamos

```python
def u_0(x,N=20): # N es el número de armónicos
    f = zeros(len(x))
    for n in range(1,N):
        f += (1-cos(n*pi/2))*sin(n*x)/(n)
    return (2/pi)*f
```

```python
f_n = u_0(x)
#print(f_n[:10])
plot(x,f_n,label='$f(x)$ con $N=20$')
plot(x,f,label='$g(x)$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Función definida a trozos y Serie de Fourier de la misma")
xlim([-0.01,2.1]),ylim([-1.5,2.2]),grid(),legend()
show()
```

```python
for i in range(20,100,20):
    plot(x,u_0(x,i),label='$g(x)$ con $N=%d$' %i)
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Serie de Fourier con diferentes armónicos")
xlim([-0.01,2.01]),ylim([-2.01,2.2]),grid(),legend()
show()
```
